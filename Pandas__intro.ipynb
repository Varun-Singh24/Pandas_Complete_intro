````python
titanic[["Sex", "Age"]].groupby("Sex").mean()
````

````python
titanic.groupby("Sex")["Age"].mean()
````

````python
# splitting the data frame into groups
male_titanic = titanic [["Age", "Sex"]] [titanic ["Sex"] == "male"]
female_titanic = titanic [["Age", "Sex"]] [titanic ["Sex"] == "female"]

# applying a function
male_age_avg = male_titanic ["Age"].mean()
female_age_avg = female_titanic ["Age"].mean()

# combining the result
age_by_sex = pd. DataFrame({"Sex": ["female", "male"], "Age": [female_age_avg, male_age_avg]})
age_by_sex
````
